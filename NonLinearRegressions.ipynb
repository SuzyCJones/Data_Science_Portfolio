```r
library("tidyr")
library("ggplot2")
library("dplyr")
```

```r
#Graphing the quadratic relationship
```

```r
xy1 <- ggplot(nonlinear, aes(x = X1, y = Y1)) +
  geom_point() + stat_smooth(method = "lm", formula = y ~x + I(x^2), linewidth =1)
```

```r
xy1
```

```r
xy2 <- ggplot(nonlinear, aes(x = X2, y = Y2)) +
  geom_point() + stat_smooth(method = "lm", formula = y ~x + I(x^2), size =1)
xy2
```

```r
# X1 and Y1 is a quadratic relationship
# X2 and Y2 is an exponential relationship
# Miss Torin (the three-year-old) would like to add to the analysis that they are opposites. 
```

```r
# quadratic squared
xsq <- nonlinear$X1^2
```

```r
#linear regression model for quadric
quadModel <- lm(nonlinear$Y1 ~ nonlinear$X1+xsq)
summary(quadModel)
```

```r
#The F-statistic and p-value render the model significant, the null hypothesis is valid
```

```r
# linear regression model for exponential
exMod <- lm(log(nonlinear$Y2)~nonlinear$X2)
summary(exMod)
```

```r
#The F-statistic and p-value render the model significant as well, the null hypothesis is valid
```
